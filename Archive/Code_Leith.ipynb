```python
def calculate_sharpe_ratio(df):
    
    df_pct_change = df.pct_change()
    #daily_return
    sharpe_ratio = df_pct_change['close'].mean() / df_pct_change['close'].std()
    
    #annualized sharpe ratio
    return (252 ** 0.5) * sharpe_ratio
```

```python
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")

# Format current date as ISO format
start_date = pd.Timestamp("2011-05-01", tz="America/New_York").isoformat()
today = pd.Timestamp("2020-12-02", tz="America/New_York").isoformat()


# Set the tickers
tickers = ["SPY", "QQQ", "VTI",
           "SPBO", "JNK", "LQD",
           "IYR", "VNQ", "USRT",
           "PSP", "USO", "GLD",
           "MUB", "MUNI", "PZA",
           "VMBS", "MBB", "MBG",
           "TLT", "SCHO", "SHY"
          ]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get closing prices
df_ETFs = alpaca.get_barset(
    tickers,
    timeframe,
    start = start_date,
    end = today
).df

df_ETFs.head()
```

```python
sharpe_ratio_list = []
# for each of the tickers, we will calculate the sharpe ratio and store it in a list of dictionaries 
for ticker in tickers:
    sharpe_ratio = calculate_sharpe_ratio(df_ETFs[ticker])
    sharpe_ratio_list.append({"ticker": ticker, "sharpe_ratio" : sharpe_ratio})

# create a data frame with from the list
df_sharpe_ratio= pd.DataFrame(sharpe_ratio_list)
df_sharpe_ratio
```

```python
# Create and empty DataFrame for closing prices
df_closing = pd.DataFrame()

# Fetch the closing prices of KO and TSLA
df_closing["SPY"] = df_ETFs["SPY"]["close"]
df_closing["QQQ"] = df_ETFs["QQQ"]["close"]
df_closing["VTI"] = df_ETFs["VTI"]["close"]
df_closing["SPBO"] = df_ETFs["SPBO"]["close"]
df_closing["JNK"] = df_ETFs["JNK"]["close"]
df_closing["LQD"] = df_ETFs["LQD"]["close"]
df_closing["IYR"] = df_ETFs["IYR"]["close"]
df_closing["VNQ"] = df_ETFs["VNQ"]["close"]
df_closing["USRT"] = df_ETFs["USRT"]["close"]
df_closing["PSP"] = df_ETFs["PSP"]["close"]
df_closing["USO"] = df_ETFs["USO"]["close"]
df_closing["GLD"] = df_ETFs["GLD"]["close"]
df_closing["MUB"] = df_ETFs["MUB"]["close"]
df_closing["MUNI"] = df_ETFs["MUNI"]["close"]
df_closing["PZA"] = df_ETFs["PZA"]["close"]
df_closing["VMBS"] = df_ETFs["VMBS"]["close"]
df_closing["MBB"] = df_ETFs["MBB"]["close"]
df_closing["MBG"] = df_ETFs["MBG"]["close"]
df_closing["LTL"] = df_ETFs["TLT"]["close"]
df_closing["SCHO"] = df_ETFs["SCHO"]["close"]
df_closing["SHY"] = df_ETFs["SHY"]["close"]

# Drop the time component of the date
df_closing.index = df_closing.index.date

df_closing.head()
```

```python
# Use the `pct_change` function to calculate daily returns
daily_returns = df_closing.pct_change()
daily_returns.head()
```

```python
# Use the `std` function and multiply by the square root of the number of trading days in a year to get annualized volatility
volatility = daily_returns.std() * np.sqrt(252)
volatility.sort_values(inplace=True)
volatility
```
